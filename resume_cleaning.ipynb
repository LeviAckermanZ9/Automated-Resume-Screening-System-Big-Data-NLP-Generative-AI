```python
df["Skills"] = df["Skills"].str.lower()
df["Job_Role"] = df["Job_Role"].str.lower()

df
```

```python
import re

df["Skills"] = df["Skills"].apply(lambda x: re.sub(r'[^\w\s]', "", x))
df["Job_Role"] = df["Job_Role"].apply(lambda x: re.sub(r"[^\w\s]", "", x))
df
```

```python
import nltk

nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize as wt

skills_tokens = []
for value in df["Skills"]:
    skills_tokens.append(wt(value))

df["Skills_Tokens"] = skills_tokens

job_role_tokens = []
for value in df["Job_Role"]:
    job_role_tokens.append(wt(value))

df["Job_Role_Tokens"] = job_role_tokens

df
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

df["Skills_Tokens"] = df["Skills_Tokens"].apply(lambda token: [word for word in token if word not in stop_words])
df["Job_Role_Tokens"] = df["Job_Role_Tokens"].apply(lambda token: [word for word in token if word not in stop_words])

df
```
